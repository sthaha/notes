# Filter by Rows

* use `loc`
* use `iloc`

# 

```python
df.iloc[1, 3] # whats at row 1, col 3 (0 indexed)
```

```python
df.iloc[:5, ] # every thing from 0-5, and all cols
```

```python
df.iloc[5] # single row all cols 
```

```python
df.iloc[5, :8] # single row only 8 cols 
```

```python
df.iloc[[5]] # single row as a data frame
```

```python
df.iloc[:, [3, 5]] # all row of col 3, 5
```

### Using loc (for names)

```python
df.loc[:, ["Airline", "Origin"]]
```
